```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
ecarttype = 10
N = 1000
exp = np.random.exponential(1,N)
interval = np.linspace(0,5,100)

plt.hist(exp, bins = interval,density=True,alpha=0.8,)

densite = stats.expon().pdf
plt.plot(interval,densite(interval))
plt.legend([f'densitée',f'échantillon'])
plt.grid()
plt.show()
```

```python
for N in [1_000,10_000,100_000]:
    X,Y = 2 * np.random.rand(N) - 1,2 * np.random.rand(N) - 1
    interval2 = np.linspace(-3,3,100)

    plt.hist(X+Y, bins = interval2,density = True,alpha=0.8)

    densite2 = stats.uniform(-1,2).pdf
    plt.plot(interval2,densite2(interval2))

    x = np.linspace(-1,1,N)
    plt.plot(x,1/4 * np.maximum(2 - abs(x),0))
    plt.legend([f'densitée',f'échantillon'])
    plt.grid()
    plt.show()
```

```python
N = 10_000
X, Y, Z = np.random.exponential(1,N),np.random.exponential(1,N)/2,np.random.exponential(1,N)/2 
interval3 = np.linspace(-3,3,100)

plt.hist(X - Y, bins = interval3,density=True,alpha=0.8)
plt.hist(Z, bins = interval3,density=True,alpha=0.8)

densite3 = stats.expon().pdf
plt.plot(interval3,densite3(interval3))
plt.legend([f'densitée',f'échantillon'])
plt.grid()
plt.show()
```

```python
N = 1000
p = 1/2
plt.figure(figsize=(14,6))
for i,n in enumerate([10,100,1000]):
    plt.subplot(1,3,1+i)
    Y = np.random.binomial(n,p,N)
    interval4 = np.linspace(np.min(Y),np.max(Y),n)
    plt.hist(Y, bins = interval4,density = True,alpha=0.8)
    x = np.arange(stats.binom.ppf(0.0001, n, p),stats.binom.ppf(0.99999, n, p))
    print(np.max(x),np.max(Y))
    plt.plot(x, stats.binom.pmf(x,n,p))
    plt.legend(['répartition binomiale',f'échantillon de X pour n = {n}'],fontsize=8)
    plt.grid()
```

```python
def uniforme(N):
    return 2 * np.random.rand(N) - 1
```

```python
nb_rep = 100_000
nb_lgn = 1000
list_Sn = []

for _ in range(N):
    X = uniforme(nb_rep)
    list_Sn.append(np.mean(X))
    
nb_bins = 100
interval = np.linspace(np.min(list_Sn),np.max(list_Sn),nb_bins)
plt.hist(list_Sn,bins = interval,density = True,alpha=0.8)
plt.grid()
plt.show()

plt.hist(list_Sn,bins = np.linspace(-1,1,nb_bins),density = True,alpha=0.8)
plt.grid()
plt.show()
```

```python
nb_rep = 100_000
nb_lgn = 1000
list_Sn = []

for _ in range(N):
    X = uniforme(nb_rep)
    list_Sn.append(np.mean(np.tan(np.pi*X/2)))
    
nb_bins = 100
interval = np.linspace(-10,10,nb_bins)
plt.hist(list_Sn,bins = interval,density = True,alpha=0.8)
plt.grid()
plt.show()
```
