**Exercice 3 (exponentielle de matrice)**

Soit $A$ une matrice carrée. Nous définissons l'exponentielle de la matrice $A$, notée $e^A$ la matrice définie par la somme infinie
$$ e^A = \sum_{n=0}^\infty \frac{1}{n!} A^N.$$

> - Proposez une fonction `exp_matrice` qui prend en argument une matrice `A`, qui vérifie que cette matrice est carrée et qui retourne une valeur approchée de l'exponentielle de `A`.
> - Vérifiez que votre fonction est correcte en calculant les exponentielles de ces trois matrices :
\begin{equation*}
A_1 = \begin{pmatrix} 1&0 \\ 0&2 \end{pmatrix}, \quad
A_2 = \begin{pmatrix} 1&1 \\ 0&1 \end{pmatrix}, \quad
A_3 = \begin{pmatrix} 1&-1 \\ 1&2 \end{pmatrix}.
\end{equation*}
Vous devez trouver des résultats proches de ceux-ci :
```
[[2.71828183 0.        ]
 [0.         7.38905607]]
 
[[2.71828183 2.7182818 ]
 [0.         2.71828183]]
 
[[ 0.93244685 -3.94211457]
 [ 3.94211457  4.87456142]]
```

*Indication : vous calculerez tous les termes $A^n/n!$ jusqu'à ce que ce terme soit plus petit qu'une précision donnée (vous pourrez ajouter cette précision en argument optionnel à votre fonction). La matrice $A^n/n!$ sera dite petite si `np.linalg.norm(A^n/n!) < epsilon`.*



````python
def exp_matrice(A,ε):
    if A.ndim == 2:
        An = A + np.eye(A.shape[0])
        un = A 
        n = 1
        if A.shape[0] == A.shape[1]:
            while np.linalg.norm(un) > ε :
                un = A/(n+1) @ un  
                An += un 
                n += 1
        else:
            raise Value.Error ("La matrice n'est pas carré")
    else:
        raise Value.Error ("La matrice n'est pas de dimention 2")
    return An   
````

````python
A1 = np.array([[1,0],[0,2]])

exp_matrice(A1,1e-6)
````

````python
A2 = np.array([[1,1],[0,1]])
exp_matrice(A2,1e-6)
````

````python
A3 = np.array([[1,-1],[1,2]])
exp_matrice(A3,1e-6)
````

Nous allons faire un peu de géométrie du plan dans cet exercice. Un point du plan sera assimilé à un tableau de deux réels, correspondants à l'absisse et à l'ordonnée.

Etant donné un point $(x, y)$, la rotation $r_\theta$ d'angle $\theta$ autour du point $(0, 0)$ est définie par
$$
r_\theta
\begin{pmatrix}x\\ y\end{pmatrix} = \begin{pmatrix}\cos\theta&-\sin\theta\\ \sin\theta&\cos\theta \end{pmatrix}\begin{pmatrix}x\\ y\end{pmatrix} = \begin{pmatrix}x\cos\theta-y\sin\theta\\ x\sin\theta+y\cos\theta\end{pmatrix}.
$$
La rotation $r_\theta^{c_x,c_y}$ d'angle $\theta$ autour du point $(c_x,c_y)$ est définie par
$$
r_\theta^{c_x,c_y}
\begin{pmatrix}x\\ y\end{pmatrix} = \begin{pmatrix}c_x\\ c_y\end{pmatrix} + \begin{pmatrix}\cos\theta&-\sin\theta\\ \sin\theta&\cos\theta \end{pmatrix}\begin{pmatrix}x-c_x\\ y-c_y\end{pmatrix} = \begin{pmatrix}c_x + (x-c_x)\cos\theta-(y-c_y)\sin\theta\\ c_y + (x-c_x)\sin\theta+(y-c_y)\cos\theta\end{pmatrix}.
$$
